```python
import numpy as np

# Definir la matriz original
A = np.array([[3, -5, 6],
              [1, -1, -1],
              [8,  9, 15]])

# Calcular la inversa
A_inv = np.linalg.inv(A)

# Multiplicaci√≥n A * A_inv
resultado = np.dot(A, A_inv)

print("Matriz original A:")
print(A)

print("\nInversa de A:")
print(A_inv)

print("\nComprobaci√≥n A * A_inv:")
print(resultado)

# Opcional: comparar con la identidad
print("\nMatriz identidad esperada:")
print(np.eye(3))
```

```python
x = np.linspace(0, 1*np.pi, 10000)
y1 = np.sin(x)
y2 = np.sin(6*x)
y_total = y1 + y2

plt.plot(x, y1, label="Onda 1 (k)")
plt.plot(x, y2, label="Onda 2 (2k)")
plt.plot(x, y_total, label="Superposici√≥n", linewidth=2)
plt.legend()
plt.title("Superposici√≥n de ondas")
plt.show()
```

```python
X, Y = np.meshgrid(np.linspace(0, 5, 50), np.linspace(0, 5, 50))
Z = np.sin(2*np.pi*X) * 5*np.sin(np.pi*Y)

plt.imshow(Z, extent=(0,5,0,5), origin="lower", cmap="viridis")
plt.colorbar(label="Amplitud")
plt.title("Onda estacionaria en 2D")
plt.show()
```

```python
# @title Texto de t√≠tulo predeterminado
import numpy as np

# Definir la matriz original
A = np.array([[3, -5, 6],
              [1, -1, -1],
              [8,  9, 15]])

# Calcular la inversa
A_inv = np.linalg.inv(A)

# Multiplicaci√≥n A * A_inv
resultado = np.dot(A, A_inv)

print("Matriz original A:")
print(A)

print("\nInversa de A:")
print(A_inv)

# Redondear el resultado a 3 decimales
resultado_redondeado = np.round(resultado, decimals=3)

print("Matriz A * A_inv redondeada:")
print(resultado_redondeado)

# Opcional: comparar con la identidad
print("\nMatriz identidad esperada:")
print(np.eye(3))

```
